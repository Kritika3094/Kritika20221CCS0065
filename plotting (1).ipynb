```python
import matplotlib.pyplot as plt
x=[1,2,4,4]
y=[1,4,4,4]
plt.plot(x,y,color="r")
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[10,20,30,40]
mylabels=["Apple","Banana","Grape","Dates"]
mycolors=["pink","lavender","purple","skyblue"]
d=[0.2,0.3,0.4,0.6]
plt.pie(x,labels=mylabels,colors=mycolors,explode=d, shadow=True)
plt.show()
```

```python
x=[3,6,8,4]
y=[6,7,8,9]
plt.title("bar plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.bar(x,y)

```

```python
x=[1,3,5,7]
y=[2,4,6,8]
colors=["g","r","b","orange"]
plt.bar(x,y,color=colors,)
plt.xlabel("X axis", fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.title("Bar plot")
plt.show()
```

```python
x=[1,3,5,7]
y=[2,4,6,8]
colors=["red","yellow","blue","brown"]
sizes=[500,600,700,800]
plt.scatter(x,y,color=colors,s=sizes,alpha=0.9,marker="*",edgecolor="r")
plt.xlabel("X axis", fontsize=10)
plt.ylabel("Y axis",fontsize=10)
plt.title("Scatter plot",fontsize=15)
plt.show()
```
