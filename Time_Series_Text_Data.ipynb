```python
import pandas as pd
import matplotlib.pyplot as plt

data={"date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01"]),
                            "value":[10,15,22,28,35]}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("date")
plt.ylabel("value")
plt.title("time series with trend")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data={"date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2023-03-01","2023-04-01"]),
                            "value":[5,10,20,30,15,12,20,30]}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("date")
plt.ylabel("value")
plt.title("Time series with Seasonality ")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
text_data="Data visualization is the graphical representation of information and data in a pictorial or graphical format (Visualization of Data could be: charts, graphs, and maps). Data visualization tools provide an accessible way to see and understand trends, patterns in data, and outliers. Data visualization tools and technologies are essential to analyzing massive amounts of information and making data-driven decisions. The concept of using pictures is to understand data that has been used for centuries"
wordcloud=WordCloud(width=800,height=600,background_color='White').generate(text_data)
plt.figure(figsize=(10,15))
plt.imshow(wordcloud)
plt.title('word cloud')
plt.axis('off')
plt.show()
words=text_data.split()
word_counts=Counter(words)
plt.figure(figsize=(10,15))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('histogram of words frequency')
plt.xlabel('words')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show
```
